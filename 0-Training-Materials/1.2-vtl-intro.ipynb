## A brief introduction <img style="float: right;" src="images/VTL_13.png" width="200" height="300">

**Validation & Transformation Language (VTL)** is a standard scripting language for defining validation and transformation rules on statistical data.

It is used in data integration and statistical production to:
- **validate** data quality (rules, consistency checks)
- **transform** data (derive new data sets, filter, reshape columns…)

VTL is **declarative and functional**: you describe *what* should be computed, not *how* the engine should execute it step by step.

In this training, VTL scripts run inside **Trevas Jupyter** (see [1.1-trevas-intro.ipynb](1.1-trevas-intro.ipynb)).


## Why use VTL? <img style="float: right;" src="images/VTL_14.png" width="300" height="300">

VTL is designed for **users in business/statistical domains**, not only for IT developers.

| Property | Benefit |
|---|---|
| User-oriented information model | Tables, identifiers, measures — familiar concepts |
| Mathematical style | Expressions like `result := A + B` |
| English-based syntax | Readable and shareable across countries |
| Standard operators | Same semantics across implementations (SDMX / GSIM ecosystem) |
| Data-production focus | Operators for validation, joins, aggregation, time series… |

A transformation in VTL is always an **assignment**:

`result := expression`

or, for a result that should be stored persistently:

`result <- expression`


## Self-service vs IT support <img style="float: left;" src="images/VTL_20.png" width="400" height="400">

**Business users** can write and maintain VTL rules for validations and transformations that match their domain knowledge.

**IT** provides and operates:
- the VTL engine (Trevas) and runtime (Spark, storage, security)
- performance, scalability, and reliability
- integration with source systems and dissemination pipelines

You focus on the **logic**; the engine focuses on **execution**.


## IT support <img style="float: right;" src="images/VTL_1.png" width="300" height="300">

**Business users** define rules and transformations against their domain knowledge.

**IT** takes care of the rest:
- maintaining and extending the VTL engine that interprets scripts
- performance and scalability for large data volumes
- reliability, fault tolerance, and security

You concentrate on **what** the data should look like; IT ensures **how** it runs at scale.


## VTL language elements (physical view) <img style="float: right;" src="images/VTL_17.png" width="30" height="30">

A **data set** is a table: rows are **data points**, columns are **components**.

| Concept | Meaning |
|---|---|
| **Identifier** | Independent variable — defines *which* observation (e.g. student id, country, period) |
| **Measure** | Dependent variable — the value being counted or observed (e.g. grade, amount) |
| **Attribute** | Descriptive context attached to an observation (e.g. school name, teacher) |
| **Data structure** | The list of components **and their roles** (not only column names) |
| **Scalar** | A single value (VTL can also produce scalar results) |

Identifiers and measures/attributes play different roles in operators — this is central to VTL and covered in detail in **2-core-rules**.


## Example — Students grades (physical data)

Dataset **`students_grades.csv`** (in `data/`):

![Students grades table](images/VTL_4.png)

| Student_Id | Subject | Semester | Grade | School | Teacher |
|---|---|---|---|---|---|
| 001 | Math | Fall | 90.5 | ABC School | Mr. Smith |
| 001 | English | Fall | 85.0 | ABC School | Mrs. Johnson |
| 002 | Math | Fall | 88.0 | XYZ School | Mrs. Davis |

The **same values** can be described logically as:

| Role | Components |
|---|---|
| Identifiers | Student_Id, Subject, Semester |
| Measure | Grade |
| Attributes | School, Teacher |


## Information model (logical view) <img style="float: right;" src="images/VTL_8.png" width="30" height="30">

VTL manipulates **logical data sets**, not only CSV files.

```
┌─────────────────────────────────────────────┐
│  Data set                                   │
│  ┌──────────── identifiers ────────────┐    │
│  │ Id_1 │ Id_2 │ …                     │    │
│  ├──────────── measures / attributes ──┤    │
│  │ Me_1 │ Me_2 │ At_1 │ …              │    │
│  └─────────────────────────────────────┘    │
│  rows = data points                         │
└─────────────────────────────────────────────┘
```

**Key idea:** measures and attributes are *dependent* — they acquire business meaning only together with the identifiers of the observation.

The engine uses **roles** when applying operators (for example `keep` and `drop` behave differently on measures than on identifiers). We explore this in the next notebooks.
