## Load a SAS dataset

```vtl
taxi := loadSas("data/nyc_taxi.sas7bdat");
```

## Inspect data

```vtl
m := showMetadata(taxi);
d := show(taxi);
```

## Define VTL roles and keep useful columns

```vtl
taxi_meta := taxi  [calc identifier vendor_id := vendor_id, identifier pickup_at := pickup_at]
                            [keep total_amount];
```

## Compute `ratio_to_report`

```vtl
ratio <- ratio_to_report(taxi_meta over(partition by vendor_id)) * 1000000;
```

## Write parquet file

```vtl
taxi := writeParquet("output", ratio);
```
