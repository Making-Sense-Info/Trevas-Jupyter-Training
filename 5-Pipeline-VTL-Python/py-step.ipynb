## Read the file produced with VTL and save 10 first rows

```python
import glob
from pathlib import Path

import pandas as pd

INPUT_GLOB = "vtl-step-output/part-*.csv"
OUTPUT_DIR = Path("py-step-output")
OUTPUT_FILE = OUTPUT_DIR / "top10.csv"

parts = sorted(glob.glob(INPUT_GLOB))
if not parts:
	raise FileNotFoundError(f"No CSV parts found at {INPUT_GLOB} (run the VTL step first)")

df = pd.concat((pd.read_csv(f) for f in parts), ignore_index=True)
top10 = df.head(10)

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
top10.to_csv(OUTPUT_FILE, index=False)

print(f"Wrote {len(top10)} rows to {OUTPUT_FILE}")
```
