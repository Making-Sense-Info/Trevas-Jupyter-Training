## A brief introduction <img style="float: right;" src="images/VTL_13.png" width="200" height="300">


Validation & Transformation Language (VTL) is a scripting language used in data integration tasks. It’s commonly utilized in data warehouse environments to manipulate data and ensure its quality. 
VTL is a standard language for defining validation and transformation rules (set of operators, their syntax and semantics) for any kind of statistical data.
The range of application fields for VTL is broad, given the universal business need for reliable, high-quality data.

In this interactive tutorial, we aim to introduce the usage of VTL, elaborate on its benefits and delve deeper into its areas of application.

With Laura, we guide you through this tutorial.

This language is designed for **users without information technology (IT) skills by means all users of business areas**. Laura as a business manager is e.g., responsible for managing metadata that drive collection, compilation and dissemination processes.  
Since the VTL is **based on a “user” perspective and “user” information model**, it is intuitive and friendly with the goal that users should be able to define and understand validations and transformations as easily as possible.

## VTL Language Elements <img style="float: right;" src="images/VTL_17.png" width="30" height="30">  

Laura wants to define quality business rules to check data stored in a table-like structure. For this, she needs to know the language elements that she can use. 
Let’s get some insides into the language details to understand VTL. It all starts from a table ...

- *Data set:* A table that consists of various columns that have different meanings.
- *Identifier:* A column that provides a number, date or text with specific business meaning.
- *Measure:* A column that provides a number, date or text with specific business meaning.
- *Attribute:* A column that provides additional context to another column in a table.
- *Set:* A list of allowed values to be used in a column.
- *Subset:* Some values from the overall list of allowed values.
- *Data Point:* A record within a table.
- *Domain:* Different types of lists can be created to group similar items. Each group will get a common name.
- *Component:* Part of a table, an identifier column, a measure column, etc. \n
- *Data Set component:* List of values within one selected component. It can consist of a subset of allowed values of a Domain for this Data set, e.g., all countries mentioned in a column named Residence.
- *Scalar:* A value in a column selected within a record.
- *Data Structure:* It is a data set that holds beside the column names also the information about the type of the column (identifier, measure, attribute).

## General structure of VTL statements <img style="float: right;" src="images/VTL_21.png" width="100" height="30">

A statement in VTL is a single, independent unit of code that performs a specific action. As mentioned above, the statement is terminated by a semicolon. Each statement in VTL is an assignment statement and starts (on the left-hand side) with a variable that holds the result. After the variable, an assignment operator follows like “:=”. The right-hand side holds the action to be evaluated. 

VTL distinguishes between temporary and persistent assignments. Persistent assignments (**<-**) are expected to be stored in a database. Temporary assignments (**:=**) are used for intermediate results.

# Join Operators Types <img style="float: right;" src="images/VTL_23.png" width="100" height="80">

Laura wants to *explore the Operators* that are introduced in VTL language. Let's explore them together!

VTL is made of Operators, which are the basic operations that the language can do. For example, the VTL has some mathematical operators (e.g. sum(+), subtraction (-), multiplication (* ), division (/), string operatores (e.g. string concatenation, substring...), comparison operatios (e.g. equal(=), greater than (>), lesser than (<), ...), logical operators and so on.

The core set of operations; they are the primitive operations, in the sense that all the other operations can be defined in their terms. 

<u> The core operations are: </u>  
- The operations that **accept scalar arguments as operands and return a scalar  value** (for example the sum between numeric scalar values, the  concatenation between string scalar values, a logical operation between  boolean scalar values …).
- The various kinds of **Join operators**, which allow to lift the scalar operations to the Data Set level, i.e., to compose  Data Sets with scalar values or with  other Data Sets.
- **Other special operators** which cannot be defined by means of the previous two categories (for example the analytical functions).  

*What are Join Operators?*

The Join operators are **fundamental VTL operators**. They are part of the core of the language and allow to **obtain the behaviour of the majority of the other non-core operators**, plus many additional behaviours that cannot be obtained through the other operators.
The Join operators are **four, namely the inner_join, the left_join, the full_join and the cross_join**. Because their syntax is similar, they are described together.The more general and powerful behaviour is supplied by the Join operators, which operates at **Data Set level and allow to compose one or more Data Sets** in many possible ways. 


<u>   </u>


<u> In particular, the Join operators allow Laura to: </u> 




<img style="float: right;" src="images/VTL_24.png" width="400" height="80"> 

- **match** the Data Points of the input Data Sets by means of various matching options (inner, left, full, cross_join) and by specifying the Components to match ("using" clause). For example the sentence *inner_join_ D1, D2 using [ reference_date, geo_area ]*
- **filter** the result of the match acoording to a condition, for example the sentence *filter  D1#M1  >  0* maintains the matched Data Points for which the Measure M1 of D1 is positive.
- **aggregate** according to the values of some Identifier, for example the sentence *group by [ Id1 , Id2 ]* eliminates the Identifiers save than Id1 and Id2 and aggregate the Data Points having 2879 the same values for Id1 and Id2 
- **combine** homonym measures of the matched Data Points according to a formula, for example the sentence *apply  D1 + D2* sums the homonymous Measures of the matched Data Points of D1 and D2
- **calculate** new Components (or new values for existing Components) according to the desired formulas, also assigning or changing the Component role (Identifier, Measure, Attribute), for example: *calc  measure  M3 := M1 + M2 , attribute A1 := A2 || A3* calculates the Measure M3 and the Attribute A1 according to the formulas above
- **keep or drop** the specified Measures or Attributes, for example the sentence *keep  [M1 , M3,  A1]* maintains only the specified measures and attributes, instead the sentence *drop  [M2 ,  A2,  A3 ]* drops only the specified measures and attributes
- **rename** the specified Components, for example:  *rename  [M1 to M10 ,  I1 to I10]*



*Let's look at the syntax of Join Operators:* 

Join is defined either as <b><font> <font color=#003398>inner_join</font></b>, <b><font> <font color=#407888>left_join</font></b>, <b><font> <font color=#003399>full_join</font></b>,<b><font> <font color=#407888>cross_join</font></b>

```vtl
/* 
joinOperator ( ds { as alias } { , ds { as alias } }* { using usingComp { , usingComp }* }
{ filter filterCondition }
{ apply applyExpr 
| calc calcClause 
| aggr aggrClause { groupingClause } } 
{ keep comp {, comp }* | drop comp {, comp }* }

{ rename compFrom to compTo { , compFrom to compTo }* }
)
*/
```

**Semantics for scalar operations:**

This replaces all the occurrences of a specified string-pattern (pattern1) with another one (pattern2). If pattern2 is omitted then all occurrences of pattern1 are removed. 

For example:


**Behaviour**
As for the invocations at Data Set level, the operator has the behaviour of the “Operators applicable on one **Scalar Value or Data Set or Data Set Component**”, as for the invocations at Component or Scalar level, the operator has the behaviour of the “Operators applicable on more than two Scalar Values or Data Set Components".
